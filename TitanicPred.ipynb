```python
import pandas as pd
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
print(df)
```

```python
df = df.drop("Cabin", axis = 1)
df = df.drop("Ticket", axis = 1)
df
```

```python
medianAge = df['Age'].median()
df.fillna({'Age':medianAge}, inplace = True)

print(df.sample(10))
```

```python
df['Sex']= df['Sex'].map({'female': 0, 'male': 1})
df = df.drop('Name', axis = 1)

df
```

```python
modus_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(modus_embarked, inplace = True)

df
```

```python
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
df
```

```python
df.info()
```

```python
z = df.drop(columns = ['Embarked_Q', 'Embarked_S'], axis = 1)
z
```

```python
import seaborn as sns
sns.pairplot(z, hue = 'Survived', height = 3)
```

```python
titanic_sum = df.select_dtypes(include= ['float64', 'int64'])
titanic_cor = titanic_sum.corr()['Survived'].drop('Survived')
top_features = titanic_cor[abs(titanic_cor) > 0.2].sort_values(ascending=False) #displays pearsons correlation coefficient greater than 0.5

print(f'This is {len(top_features)} strongly correlated values with Survived = \n{top_features}')
```

```python
df['Sex'].value_counts()
```

```python
x = df.drop('Survived', axis = 1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split

# test_size=0.2 berarti 20% data untuk testing
# random_state=42 untuk memastikan hasil split selalu sama
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)
print('Proporsi di y_train:\n', y_train.value_counts(normalize=True))
print('\nProporsi di y_test:\n', y_test.value_counts(normalize=True))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score,StratifiedKFold, GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import matplotlib
import matplotlib.pyplot as plt
```

```python
skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 42)
```

```python
pipe = Pipeline([('ss', StandardScaler()),
                 ('rfc', RandomForestClassifier(n_estimators=200,
                                                max_depth=15,
                                                min_samples_leaf=7,
                                                random_state=42))])
```

```python
param_grid = {
    'rfc__n_estimators': [100, 200, 300],
    'rfc__max_depth': [5, 10, 15, 20, None],
    'rfc__min_samples_leaf': [1, 3, 5, 7, 9],
    'rfc__max_features': ['sqrt', 'log2', 0.5, None]
}
```

```python
grid_search = GridSearchCV(estimator = pipe,
                           param_grid = param_grid,
                           cv = skf,
                           scoring = 'f1',
                           n_jobs=-1,
                           verbose=1)
```

```python
grid_search.fit(x_train, y_train)
```

```python
print('Parameter terbaik yang ditemukan:')
print(grid_search.best_params_)
```

```python
model= grid_search.best_estimator_
print(model)
```

```python
y_pred = model.predict(x_test)
```

```python
y_prob = model.predict_proba(x_test)[:, 1]
y_prob
```

```python
from sklearn.metrics import classification_report, RocCurveDisplay, ConfusionMatrixDisplay
```

```python
print(classification_report(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_estimator(model, x_test, y_test, cmap='Blues')
plt.title("Confusion Matrix (Data Test)")
plt.show()
```

```python
RocCurveDisplay.from_predictions(y_test, y_prob)
plt.title("ROC Curve (Data Test)")
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
accuracy
```
